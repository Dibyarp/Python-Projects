```python
inp1 = app_data
inp1
```

```python
inp1['Installs'] = np.log1p(inp1.Installs)
inp1
```

```python
inp1['Reviews'] = np.log1p(inp1.Reviews)
inp1
```

```python
inp1.drop(['App', 'Last_Updated' , 'Current_Ver' , 'Android_Ver'], axis=1 , inplace=True)
inp1
```

```python
inp1.Category.unique()
```

```python
inp1
```

```python
catg = pd.get_dummies(inp1.Category, dtype='int64')
catg
```

```python
import pandas as pd
con_rat = pd.get_dummies(inp1.Content_Rating, dtype='int64')
con_rat
```

```python
gen = pd.get_dummies(inp1.Genres , dtype='int64')
gen
```

```python
inp2 = pd.concat([inp1 , catg , con_rat, gen] , axis=1)
inp2
```

```python
inp2.drop(['Category', 'Content_Rating', 'Genres'], axis =1 , inplace=True)
```

```python
inp2.dtypes
```

```python
#Model Building Starts
#Declaration of x and y
x = inp2.drop(['Rating'], axis =1)
y= inp2.Rating
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.30)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
inp2_lin = LinearRegression().fit(x_train, y_train)
```

```python
inp2_lin.intercept_
```

```python
inp2_lin.coef_
```

```python
#Model Testing
predicted_rating = pd.DataFrame(np.round(inp2_lin.predict(x_test),1), columns= ['Predicted_Rating'])
predicted_rating
```

```python
y_test
```

```python
inp2_Result = pd.concat([x_test.reset_index(drop=True),y_test.reset_index(drop=True),predicted_rating],axis=1)
inp2_Result
```

```python
diff_rating = inp2_Result[['Rating' , 'Predicted_Rating']]
diff_rating
```

```python
diff_rating['Errors'] = abs(diff_rating.Rating - diff_rating.Predicted_Rating)/diff_rating.Rating 
diff_rating
```

```python
diff_rating.Errors.mean()
```

```python
calc_accuracy = np.round((1 - diff_rating.Errors.mean()) , 2)
calc_accuracy
```

```python
from sklearn.metrics import r2_score
r_square = np.round(r2_score(diff_rating.Rating , diff_rating.Predicted_Rating), 2)
r_square
```
