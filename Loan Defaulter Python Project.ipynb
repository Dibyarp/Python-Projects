```python
# Create a dictionary to map the old column names to the new column names
new_column_names = {col: col.replace('.', '_') for col in data.columns}
new_column_names

# Rename the columns using the dictionary
data.rename(columns=new_column_names, inplace=True)

# Verify the updated column names
print(data.columns)# Retrieve the column names
column_names = data.columns.tolist()

# Convert column names to lower case
lower_case_column_names = [name.lower() for name in column_names]

# Update the column names in the DataFrame
data.columns = lower_case_column_names
data.columns
```

```python
#Rplacing Null value of Employment Type to "Un Employed"
data['employment_type'] = data['employment_type'].fillna('Unmployed')
```

```python
#Formating The Date Column
data['date_of_birth']= pd.to_datetime(data['date_of_birth'], format= '%Y-%m-%d')
data['date_of_birth' + '_Year'] = data['date_of_birth'].apply(lambda x : x.year)
data['date_of_birth' + '_Month'] = data['date_of_birth'].apply(lambda x : x.month)
data['date_of_birth' + '_Day'] = data['date_of_birth'].apply(lambda x : x.day)
data.drop('date_of_birth', axis = 1, inplace = True)
```

```python
#Formating The Date Column
data['disbursaldate']= pd.to_datetime(data['disbursaldate'], format= '%Y-%m-%d')
data['disbursal_date' + '_Year'] = data['disbursaldate'].apply(lambda x : x.year)
data['disbursal_date' + '_Month'] = data['disbursaldate'].apply(lambda x : x.month)
data['disbursal_date' + '_Day'] = data['disbursaldate'].apply(lambda x : x.day)
data.drop('disbursaldate', axis = 1, inplace = True)
```

```python
#Rearranging columns 'average_acct_age'	 into separate year an month column

data[['AVERAGE_ACCT_Yr','AVERAGE_ACCT_Month']] = data['average_acct_age'].str.split("yrs",expand=True)
data[['AVERAGE_ACCT_Month','AVERAGE_ACCT_Month1']] = data['AVERAGE_ACCT_Month'].str.split("mon",expand=True)
data["AVERAGE_ACCT_AGE"]= data["AVERAGE_ACCT_Yr"].astype(str).astype(int)+((data["AVERAGE_ACCT_Month"].astype(str).astype(int))/12)
data= data.drop(columns= ["AVERAGE_ACCT_Yr","AVERAGE_ACCT_Month",'AVERAGE_ACCT_Month1','average_acct_age'])

```

```python
#Rearranging columns 'credit_history_length' into separate year an month column

data[['CREDIT_HISTORY_LENGTH_Yr','CREDIT_HISTORY_LENGTH_Month']] = data['credit_history_length'].str.split("yrs",expand=True)
data[['CREDIT_HISTORY_LENGTH_Month','CREDIT_HISTORY_LENGTH_Month1']] = data['CREDIT_HISTORY_LENGTH_Month'].str.split("mon",expand=True)
data["CREDIT_HISTORY_LENGTH"]= data["CREDIT_HISTORY_LENGTH_Yr"].astype(str).astype(int)+((data["CREDIT_HISTORY_LENGTH_Month"].astype(str).astype(int))/12)
data= data.drop(columns= ["CREDIT_HISTORY_LENGTH_Yr","CREDIT_HISTORY_LENGTH_Month",'CREDIT_HISTORY_LENGTH_Month1','credit_history_length'])
```

```python
#Lets Drop the UniqueID column as we don't want any unique columns in the data
#Drop Categorical veriable which have Very high Unique Count

data.drop(['uniqueid','supplier_id','current_pincode_id','employee_code_id','branch_id','manufacturer_id','state_id'], axis=1 , inplace=True)
```

```python
data.head()
```

```python
import pandas as pd
catg = data[['employment_type','perform_cns_score_description']]
catg_dum = pd.get_dummies(catg)
data = pd.concat([data,catg_dum], axis =1)
data.drop(['employment_type','perform_cns_score_description'],axis=1 , inplace=True)
```

```python
data.head()
```

```python
scale = ['disbursed_amount','asset_cost','ltv']
for col in scale:
    data[col]=data[col]/data[col].mean()
data.head()
```

```python
'''for col in data_col:
    data[col]= data[col].astype('float64')'''
```

```python
x = data.drop(['loan_default'], axis=1)
y = data.loan_default
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =  train_test_split(x,y,test_size=.30, random_state=1)
```

```python
x_train
```

```python
# Model Building
from sklearn.linear_model import LogisticRegression
data_lr = LogisticRegression().fit(x_train,y_train)
```

```python
#Model Testing
predicted_rating = pd.DataFrame(np.round(data_lr.predict(x_test),1), columns= ['Predicted_Rating'])
predicted_rating
```

```python

test_result = pd.concat([y_test.reset_index(drop=True),predicted_rating], axis=1) 
test_result
```

```python
distribution = test_result['Predicted_Rating'].value_counts()
distribution
```

```python
test_result['loan_default'].value_counts()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predicted_rating))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predicted_rating)
```

```python
from sklearn.metrics import accuracy_score
accuracy= np.round(accuracy_score(y_test,predicted_rating)*100,2)
print('The Accuracy is: ',accuracy ,'%')
```
